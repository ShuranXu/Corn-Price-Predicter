# Daily Corn Price Forecasting using Linear Regression

```python
# Project 3      
#importing libraries 
import pandas as pd
import numpy as np
import seaborn as sns 
import datetime 
import plotly.express as px
import os
import matplotlib.pyplot as plt
from pathlib import Path
%matplotlib inline
import hvplot.pandas
from sklearn.linear_model import LinearRegression
```

## 1. Data Import and Cleanup

```python
# Read in corn actual price and sort by date
corn_act_prices = pd.read_csv(Path("../Resources/combined_portfolios_df.csv"), index_col="Date", infer_datetime_format=True, parse_dates=True).sort_index()
# Dropping Natural Gas and Corn Future
corn_act_prices = corn_act_prices.drop(columns=["Natural_Gas_future", "Corn_future", "USD_Index", "Ethanol_future"])
corn_act_prices.head(2), corn_act_prices.tail(2)
```

```python
# Create a series using "Corn_Price" price percentage returns, drop any NaNs, and check the results:
# (Make sure to multiply the pct_change() results by *100)
corn_act_prices['Return'] = corn_act_prices.Corn_Price.pct_change() * 100
```

```python
# Drop nulls from Corn DataFrame
corn_act_prices.dropna(inplace=True)
```

```python
# Plotting Corn Price and the Daily % Chage in the Corn Price
corn_act_prices.Corn_Price.plot(); corn_act_prices.Return.plot()
```

## 2. Prepare the Features (X) and the Target (y)

## Linear Regression with Train/Test split

```python
# Dropping Corn Price
corn_act_prices = corn_act_prices.drop(columns=["Corn_Price"])
# Train Test Split of 85% 15%
train = corn_act_prices.loc['2005-03-28':'2017-05-31']
test = corn_act_prices.loc['2017-06-1':]
train.shape, train.tail(2), test.shape, test.head(2)
```

```python
# Create dataframes for X_train and X_test
# Create series for y_train and y_test
# X_train (training set using just the independent variables), X_test (test set of of just the independent variables)
# Y_train (training set using just the "y" variable, i.e., "Futures Return"), Y_test (test set of just the "y" variable):
X_train = train.drop(columns=["Return"])
X_test = test.drop(columns=["Return"])
y_train = train["Return"]
y_test = test["Return"]
```

```python
#Fit a SKLearn linear regression using  just the training set (X_train, Y_train):
model = LinearRegression()
model.fit(X_train, y_train)
print(model.coef_)
print(model.intercept_)
```

```python
# pd.DataFrame(model.coef_, X_train.columns, columns = ['Coeff'])
```

## 3. Make predictions using the Testing Data
Note: We want to evaluate the model using data that it has never seen before, in this case: X_test.

```python
# And make a prediction of "y" values for just the test dataset
predictions = model.predict(X_test)
predictions
```

```python
# Assemble actual y data (Y_test) with predicted y data (from just above) into two columns in a dataframe:
Results = y_test.to_frame()
Results["Predicted Return"] = predictions
Results.head(3)
```

## Out-of-Sample Performance

Evaluate the model using "out-of-sample" data (X_test and y_test)

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculate the mean_squared_error (MSE) on actual versus predicted test "y" 
# A "good" MSE score will be close to zero while a "good" R2 Score will be close to 1.
mse = mean_squared_error(Results["Return"],Results["Predicted Return"])
# Using that mean-squared-error, calculate the root-mean-squared error (RMSE):
rmse = np.sqrt(mse)
# Calculate r2_score (r-squared)
r2 = r2_score(Results["Return"],Results["Predicted Return"])

# Print the r2, mser, rmse, and standard deviation
print(f"R-squared (R2 ): {r2}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Out-of-Sample Root Mean Squared Error (RMSE): {rmse}")
print(f"Standard Deviation of Futures Return: {np.std(Results.Return)}")
```

## In-Sample Performance

Evaluate the model using in-sample data (X_train and y_train)

```python
# Construct a dataframe using just the "y" training data:
in_sample_results = y_train.to_frame()

# Add a column of "in-sample" predictions to that DataFrame:  
in_sample_results["In-sample Predictions"] = model.predict(X_train)

# Calculate in-sample mean_squared_error (for comparison to out-of-sample)
in_sample_mse = mean_squared_error(in_sample_results["Return"],in_sample_results["In-sample Predictions"]
)

# Calculate in-sample root mean_squared_error (for comparison to out-of-sample)
in_sample_rmse = np.sqrt(in_sample_mse)

# Calculate r2_score (r-squared)
r2 = r2_score(in_sample_results["Return"], in_sample_results["In-sample Predictions"])

# Print the r2, mser, rmse, and standard deviation
print(f"R-squared (R2 ): {r2}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"In-sample Root Mean Squared Error (RMSE): {rmse}")
print(f"Standard Deviation of Futures Return: {np.std(in_sample_results.Return)}")
```
